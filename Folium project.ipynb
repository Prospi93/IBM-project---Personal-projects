```python
import pandas as pd 
import folium 
```

```python
df = pd.read_csv('C:\\Users\\Federico\\Downloads\\world_coordinates.csv')
```

```python
df
```

```python
mappa = folium.Map()
```

```python
segnalini = folium.map.FeatureGroup()
```

```python
for lat, lng in zip(df.latitude, df.longitude):
    segnalini.add_child(folium.features.CircleMarker(
    [lat, lng],
    radius=6,
    color='yellow',
    fill=True,
    fill_color='black',
    fill_opacity=0.8)) 
latitudes = list(df.latitude)
longitudes = list(df.longitude)
labels = list(df.Country) 
for lat, lng, labels in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=labels).add_to(mappa) 
mappa.add_child(segnalini)
```
